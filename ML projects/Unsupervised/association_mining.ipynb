```python
import pandas as pd
import numpy as np
import sklearn.preprocessing as skp
from load_data import *
from apyori import apriori
```

```python
#functions from the 12th exercise
def mat2transactions(X, labels=[]):
    T = []
    for i in range(X.shape[0]):
        l = np.nonzero(X[i, :])[0].tolist()
        if labels:
            l = [labels[i] for i in l]
        T.append(l)
    return T
def print_apriori_rules(rules):
    frules = []
    i=1
    for r in rules:
        for o in r.ordered_statistics:        
            conf = o.confidence
            supp = r.support
            x = ", ".join( list( o.items_base ) )
            y = ", ".join( list( o.items_add ) )
            print("%d.(%s) -> (%s) (supp: %.3f, conf: %.3f)"%(i,x,y, supp, conf))
            i=i+1
            frules.append( (x,y) )
    return frules
```

```python
data = load()

#dividing inputs and output
data = data.drop(['City (L/100 km)', 'Hwy (L/100 km)','Comb (mpg)','CO2 rating','Smog rating'], 1)

#scaling of the inputs
scaler = skp.StandardScaler()
scaledX = scaler.fit_transform(data)
scaledX_df=pd.DataFrame(scaledX)
scaledX_df.columns=data.columns

scaledX_df
```

```python
#binarizing the data
transformer = skp.Binarizer().fit(scaledX)
binx=transformer.transform(scaledX)

a=np.empty((1022,4))

for i in range (0,4):
    for j in range(0,1022):
        if binx[j][i]==0. :
            a[j][i]=1.
        else:
            a[j][i]=0.
binary=np.concatenate((binx, a), axis=1)
binary.shape
cols1 = ['Engine Size above', 'Cylinders above', 'Comb above', 'CO2 Emissions above','Diesel','Ethanol']
cols2 = ['RegGasoline','PremGasoline','Engine Size below', 'Cylinders below', 'Comb below', 'CO2 Emissions below']
cols=np.concatenate((cols1,cols2),axis=None).tolist()
T = mat2transactions(binary,cols)
rules = apriori( T, min_support=0.40, min_confidence=0.95)
print_apriori_rules(rules)
print('\nNice!')
```
